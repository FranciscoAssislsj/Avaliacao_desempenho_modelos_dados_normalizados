```python
house[['ExterQual','ExterCond']]
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
le.fit(house.ExterQual)
house.ExterQual = le.transform(house.ExterQual)

le.fit(house.ExterCond)
house.ExterCond = le.transform(house.ExterCond)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
house_norm = pd.DataFrame(scaler.fit_transform(house), index = house.index, columns = house.columns)
```

```python
house_norm
```

```python
house_norm.columns
```

```python
house_norm
```

```python
x = house_norm[['MSSubClass', 'LotFrontage', 'LotArea', 'OverallQual','OverallCond', 'YearBuilt', 'YearRemodAdd', 'MasVnrArea', 'ExterQual','ExterCond', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', 'TotalBsmtSF','1stFlrSF', '2ndFlrSF', 'LowQualFinSF', 'GrLivArea', 'BsmtFullBath','BsmtHalfBath', 'FullBath', 'HalfBath', 'BedroomAbvGr', 'KitchenAbvGr','TotRmsAbvGrd', 'Fireplaces', 'GarageYrBlt', 'GarageCars', 'GarageArea','WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch', '3SsnPorch','ScreenPorch', 'PoolArea', 'MiscVal', 'MoSold', 'YrSold']]

y = house_norm[['SalePrice']]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.25 , random_state = 42)
```

```python
from sklearn.datasets import make_moons
from sklearn import metrics
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, AdaBoostRegressor, GradientBoostingRegressor
```

```python
rf = RandomForestRegressor(n_estimators=1000, random_state = 42)
rf.fit(x_train, y_train)
```

```python
prediction_rf = rf.predict(x_test)

r_squared = rf.score(x, y)

print("R^2:", r_squared)
print("MAE:", metrics.mean_absolute_error(y_test, prediction_rf))
print("MSE:", metrics.mean_squared_error(y_test, prediction_rf))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, prediction_rf)))
```

```python
ada = AdaBoostRegressor(n_estimators=1000)
ada.fit(x_train, y_train)
```

```python
prediction_ada = ada.predict(x_test)

r_squared = ada.score(x, y)

print("R^2:", r_squared)
print("MAE:", metrics.mean_absolute_error(y_test, prediction_ada))
print("MSE:", metrics.mean_squared_error(y_test, prediction_ada))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, prediction_ada)))
```

```python
gbr = GradientBoostingRegressor(n_estimators=1000)
gbr.fit(x_train, y_train)
```

```python
prediction_gbr = gbr.predict(x_test)

r_squared = gbr.score(x,y)

print("R^2:", r_squared)
print("MAE:", metrics.mean_absolute_error(y_test, prediction_gbr))
print("MSE:", metrics.mean_squared_error(y_test, prediction_gbr))
print("RMSE:", np.sqrt(metrics.mean_squared_error(y_test, prediction_gbr)))
```
